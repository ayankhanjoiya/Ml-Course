### Example 1: Temperature Conversion

```python
def convert_temp(temp,unit):
    if unit == "C":
        return (temp * 9/5) + 32
    elif unit == "F":
        return (temp - 32) * 5/9
    else:
        return None 

print(convert_temp(100,"C"))
print(convert_temp(212,"F"))
```

### Example 2 : Password Strength Checker

```python
def password_check(password):
    if len(password) < 8:
        return "Password must be atleast 8 characters long"
    if not any(char.isdigit() for char in password):
        return False
    if not any(char.islower() for char in password):
        return False
    if not any(char.isupper() for char in password):
        return False
    return True

## calling the function 
print(password_check("Strongpassword123"))
```

### Example 3: Calculate the Total Cost of items in a Shopping Cart

```python
def calculate_total_cost(cart):
    total_cost = 0
    for item in cart:
        total_cost += item["price"] * item["quantity"]
    return total_cost

#Example Cart
cart = [
    {"name": "Laptop", "price": 1000, "quantity": 1},
    {"name": "Mouse", "price": 50, "quantity": 2},
    {"name": "Keyboard", "price": 80, "quantity": 1}
]
print(calculate_total_cost(cart))
```

### Example 4: Check if a string is a Palindrome

```python
def palindrome_check(word):
    cleaned_word = word.replace(" ","").lower()
    return cleaned_word == cleaned_word[::-1]

print(palindrome_check("a man a Plan a canal Panama"))
```

### Example 5: Calculate the factorial of a number using recursion

```python
def factorial(n):
    if n == 0:
        return 1 
    else: 
        return factorial(n-1) * n
    
factorial(5)
```
